### 3.Crea columnes noves (velocitat mitjana del vol, si ha arribat tard o no...).

##### Per crear la columna Velocitat_mitjana, multiplico els registres de la columna Distance per 1,60934 per aconseguir la distància en kilòmetres, no en milles. Seguidament divideixo aquest resultat entre el temps de vol entre 60. El resultat és la velocitat mitjana en km/hora. 
##### Per crear la columna que indica si el vol ha arribat tard, es té en compte que el vol no ha arribat a temps si el seu retràs és superior a 15 minuts. Aquesta columna contindrà variables booleanes. Indicará True si ha arribat tard i indicarà False si no ha arribat tard. Per fer això he fet una condició que diu si la columa ArrDelay del dataframe df_m és més gran que 15 retorna True, sino retorna False. Seguidament se li asigna la variable condition (que conté True o False) a la nova columna "Arriba_tard".

```python
df_m['Velocitat_m'] = (df_m['Distance']*1.60934)/(df_m['AirTime']/60)
condition = df_m['ArrDelay'] > 15
df_m['Arriba_tard']=condition
```
